## Linear Regression 

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
reg.fit(X_train, y_train)
```

```python
reg.score(X_test, y_test)
```

```python
from sklearn.preprocessing import StandardScaler
Scaler = StandardScaler()
```

```python
X_train_scaler = Scaler.fit_transform(X_train)
```

```python
reg.fit(X_train_scaler, y_train)
```

```python
X_test_scaler = Scaler.fit_transform(X_test)
```

```python
reg.score(X_test_scaler, y_test)
```

```python
y_pred = reg.predict(X_test_scaler)
```

## Random forest

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
forest = RandomForestRegressor()
```

```python
forest.fit(X_train_scaler, y_train)
```

```python
forest.score(X_test_scaler, y_test)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    "n_estimators" : [200],
    "max_features" : [6],
    "min_samples_split":[2]
}

grid_search = GridSearchCV(forest, param_grid, cv = 5, scoring = "neg_mean_squared_error", return_train_score = True)

grid_search.fit(X_train_scaler, y_train)
```

```python
grid_search.best_estimator_
```

```python
best_forest = grid_search.best_estimator_
```

```python
best_forest.score(X_test_scaler, y_test)
```

## Neural network

```python
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error, accuracy_score
```

```python
model = MLPRegressor(hidden_layer_sizes=(64,), activation='relu', solver='adam', max_iter=2000, learning_rate_init=0.001, random_state=42)

model.fit(X_train_scaler, y_train)

# Dự đoán giá nhà cho tập kiểm tra
y_pred = model.predict(X_test_scaler)

# Đánh giá mô hình
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error on Test Set: {mse}')
```

## KNN

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
neigh = KNeighborsRegressor(n_neighbors=4)

neigh.fit(X_train_scaler, y_train)
```

```python
neigh.score(X_test_scaler, y_test)
```

```python
param_grid = {
    'n_neighbors': [3, 4, 5, 6, 7, 8, 9],  
    'weights': ['uniform', 'distance'],
}
```

```python
grid_search = GridSearchCV(neigh, param_grid, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(X_train_scaler, y_train)
```

```python
grid_search.best_estimator_
```

```python
grid_search.best_estimator_.score(X_test_scaler, y_test)
```
